```python
img = cv2.imread("2.png")
```

```python
plt.imshow(img)
```

```python
hsv = cv2.cvtColor(img.copy(),cv2.COLOR_BGR2HSV)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT , (3,3))
closed = cv2.morphologyEx(hsv , cv2.MORPH_CLOSE , kernel )
closed = cv2.erode(closed , None , iterations=1)
closed  = cv2.dilate(closed  ,None , iterations=2)
```

```python
plt.imshow(closed)
```

```python
medianblur =cv2.medianBlur(closed,ksize = 3)
plt.imshow(medianblur)
```

```python
lower = np.array([0, 0, 127])      
upper = np.array([179, 255, 127]) 
mask = cv2.inRange(hsv , lower , upper)
#hedef_kordinatlar = []
#hedef_kordinatlar.append(cv2.boundingRect(mask)) #gri hedefin bulunduu kordinat bulunur
```

```python
plt.imshow(mask)
```

```python
closed_mask = cv2.morphologyEx(mask , cv2.MORPH_CLOSE , kernel )
closed_mask = cv2.erode(closed_mask , None , iterations=1)
closed_mask  = cv2.dilate(closed_mask  ,None , iterations=2)
```

```python
plt.imshow(closed_mask)
```

```python
cnts = cv2.findContours(closed_mask , cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_SIMPLE)
cnts
```

```python
cnts = imutils.grab_contours(cnts)
```

```python
cnts = [c for c in cnts if cv2.contourArea(c) > 80]
```

```python
hedef_kordinatlar = []
for c in cnts:
    (x,y,w,h) = cv2.boundingRect(c)
    hedef_kordinatlar.append((x,y,w,h))
```
