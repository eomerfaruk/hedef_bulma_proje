```python
img = cv2.imread("2.png")
```

```python
img_copy = img.copy()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT , (3,3))
closed = cv2.morphologyEx(img_copy , cv2.MORPH_CLOSE , kernel )
closed = cv2.erode(closed , None , iterations=1)
closed  = cv2.dilate(closed  ,None , iterations=2)
```

```python
plt.imshow(closed)
```

```python
medianblur =cv2.medianBlur(closed,ksize = 3)
plt.imshow(medianblur)
```

```python
lower = np.array([0, 0, 180])      
upper = np.array([180, 30, 220]) 
mask = cv2.inRange(img, lower , upper)
hedef_kordinatlar = []
hedef_kordinatlar.append(cv2.boundingRect(mask)) #gri hedefin bulunduu kordinat bulunur
```

```python
img = cv2.cvtColor(img.copy() , cv2.COLOR_BGR2GRAY)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT , (3,3))
closed = cv2.morphologyEx(img , cv2.MORPH_CLOSE , kernel )
closed = cv2.erode(closed , None , iterations=1)
closed  = cv2.dilate(closed  ,None , iterations=2)
medianblur =cv2.medianBlur(closed,ksize = 3)
plt.imshow(medianblur)
```

```python
threshold = cv2.threshold(img ,120, 160, cv2.THRESH_BINARY)[1]
```

```python
plt.imshow(threshold)
```

```python
cnts  = cv2.findContours(threshold , cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnts = imutils.grab_contours(cnts)
```

```python
cnts
```

```python
daire_approx_liste = []
for c in cnts:
    epsilon = 0.02 * cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, epsilon, True)

    if len(approx) >= 5 and cv2.contourArea(approx) > 10:
        daire_approx_liste.append(approx)
```

```python
daire_approx_liste
```

```python
daire_approx_liste = sorted(daire_approx_liste, key=cv2.contourArea, reverse=True)[:4]
```

```python
daire_approx_liste[0]
```

```python
img_copy = img.copy()
a,b= img_copy.shape[:2]
matris = np.ones((a,b,3) ,dtype = np.uint8)*255 
cv2.drawContours(matris , daire_approx_liste , -1 ,(0,255,0),2 )
```

```python
plt.imshow(matris)
```
